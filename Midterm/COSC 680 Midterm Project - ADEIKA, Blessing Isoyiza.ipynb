##### 1. GRAY-LEVEL CO-OCCURRENCE MATRIX (GLCM) 

```python
from skimage.feature import graycomatrix, graycoprops
import matplotlib.pyplot as plt
import glob
import numpy as np
import cv2
```

```python
# Define path to images
imgPath = "img" + '/*.jpg'

#Define Output File
glcm_result = open("csv/graylevelcm.csv", "w")


for image in glob.glob(imgPath):
    
    flower = cv2.imread(image)
    
    #Resize image
    flower = cv2.resize(flower, (591, 501), interpolation = cv2.INTER_AREA)

    #Convert image to grey 
    flower2gray = cv2.cvtColor(flower, cv2.COLOR_BGR2GRAY)

    flowerID = image[imgPath.rfind("/") + 1:]

    # Define PATCH SIZE
    PATCH_SIZE = 21
    # select some patches from flower areas of the image
    flower_locations = [(140, 37), (57, 185), (259, 603), (317, 95)]
    flower_patches = []
    for loc in flower_locations:
        flower_patches.append(flower2gray[loc[0]:loc[0] + PATCH_SIZE,
                                loc[1]:loc[1] + PATCH_SIZE])

        # select some patches from leaf areas of the image
    leaf_locations = [(49, 39), (438, 31), (428, 573), (46, 562)]
    leaf_patches = []
    for loc in leaf_locations:
        leaf_patches.append(flower2gray[loc[0]:loc[0] + PATCH_SIZE,
                                 loc[1]:loc[1] + PATCH_SIZE])
        # compute some GLCM properties each patch
    xs = []
    ys = []
    
    for patch in (flower_patches + leaf_patches):
        glcm = graycomatrix(flower2gray, distances=[5], angles=[0], levels=256, 
                            symmetric=True, normed=True)
        xs.append(graycoprops(glcm, 'dissimilarity')[0, 0])
        ys.append(graycoprops(glcm, 'correlation')[0, 0])

        # Concatenate result of glcm gray props
        glcm_props = np.concatenate([xs, ys])
    
     # write the features to file
    features = [str(f) for f in glcm_props]
    glcm_result.write("%s,%s\n" % (flowerID, ",".join(features)))
    print("Gray-Level Co-occurence Matrix: Writing result for flower image: {}".format(flowerID))
    
    
# close the index file
glcm_result.close()  
```

##### 2. LOCAL BINARY PATTERN (LBP) 

```python
from skimage import feature
import glob
import numpy as np
import cv2
```

```python
# Define the Local Binary Pattern class
class LocalBinaryPatterns:
    def __init__(self, numPoints, radius):
        #store the number of points and radius
        self.numPoints = numPoints
        self.radius = radius
        
    def describe(self, image, eps=1e-7):
        # compute the Local Binary Pattern representation
        # of the image, and then use the LBP representation
        # to build the histogram of patterns
        
        lbp = feature.local_binary_pattern(image, self.numPoints, self.radius, method="uniform")
        (hist, _) = np.histogram(lbp.ravel(), bins=np.arange(0, self.numPoints +3), 
                                 range=(0, self.numPoints + 2))
        
        # Normalize the histogram
        hist = hist.astype("float")
        hist /= (hist.sum() + eps)
        
        #Return histogram of Local Binary Patterns
        return hist


# Define path to images
imgPath = "img" + '/*.jpg'

#Define Output File
lbp_result = open("csv/lbp.csv", "w")


for image in glob.glob(imgPath):
    
    flower = cv2.imread(image)
    
    #Extract ImageID
    flowerID = image[image.rfind("/") + 1:]

    # initialize the local binary patterns descriptor
    desc = LocalBinaryPatterns(24, 8)


    #Convert image to grey 
    flower2gray = cv2.cvtColor(flower, cv2.COLOR_BGR2GRAY)

    # Apply LBP to image
    hist = desc.describe(flower2gray)

    
    lbp = hist.flatten()
    

    # write the features to file
    features = [str(f) for f in lbp]
    lbp_result.write("%s,%s\n" % (flowerID, ",".join(features)))
    print("Local Binary Pattern - Features for Flower Image: {}".format(flowerID))
    


# close the index file
lbp_result.close()
```

```python
class LocalBinaryPatterns:
    def __init__(self, numPoints, radius):
        #store the number of points and radius
        self.numPoints = numPoints
        self.radius = radius
        
    def describe(self, image, eps=1e-7):
        # compute the Local Binary Pattern representation
        # of the image, and then use the LBP representation
        # to build the histogram of patterns
        
        lbp = feature.local_binary_pattern(image, self.numPoints, self.radius, method="uniform")
        (hist, _) = np.histogram(lbp.ravel(), bins=np.arange(0, self.numPoints +3), 
                                 range=(0, self.numPoints + 2))
        
        # Normalize the histogram
        hist = hist.astype("float")
        hist /= (hist.sum() + eps)
        
        #Return histogram of Local Binary Patterns
        return hist

# Create class that calculates distance between features of two images N.B Euclidean Distance is used.
class FeatureDistCalc:
    
    def __init__(self, image1, image2):
        
        self.image1 = image1
        self.image2 = image2
        
    def colorMomentDist(self): 
        
        #Compute Mean and Standard Deviation
        (image_mean1, image_std1) = cv2.meanStdDev(self.image1)
        (image_mean2, image_std2) = cv2.meanStdDev(self.image2)
        
        #Concatenate the mean and standard deviation of image
        image_stats1 = np.concatenate([image_mean1, image_std1]).flatten()
        image_stats2 = np.concatenate([image_mean2, image_std2]).flatten()
        
        cmd = self.euclideanDistance(image_stats1,image_stats2)
        
        return cmd
    
    def colorHistogramDist(self):
        
        
        # Compute 3D Color Histogram with 8bins per channel
        image1_hist = cv2.calcHist([self.image1], [0,1,2], None, [8,8,8], [0, 256, 0, 256, 0, 256])
        image2_hist = cv2.calcHist([self.image2], [0,1,2], None, [8,8,8], [0, 256, 0, 256, 0, 256])


        #Flatten histogram
        raw_image1_hist = image1_hist.flatten()
        raw_image2_hist = image2_hist.flatten()
    
        chd = self.euclideanDistance(raw_image1_hist,raw_image2_hist)
        
        return chd
    
    def glcmDist(self):
        
        image1_gray = cv2.cvtColor(self.image1, cv2.COLOR_BGR2GRAY)
        image2_gray = cv2.cvtColor(self.image2, cv2.COLOR_BGR2GRAY)
        
        PATCH_SIZE = 21
        
        # select some patches from flower areas of the image
        image_locations = [(174, 94), (105, 382), (429, 389), (412, 178)]
        image1_patches = []
        image2_patches = []

        for flo in image_locations:
            image1_patches.append(image1_gray[flo[0]:flo[0] + PATCH_SIZE,
                                              flo[1]:flo[1] + PATCH_SIZE])
            
            image2_patches.append(image2_gray[flo[0]:flo[0] + PATCH_SIZE,
                                              flo[1]:flo[1] + PATCH_SIZE])

        # select some patches from the leaf(background) areas of the image
        bg_locations = [(34, 46), (25, 551), (451, 52), (419, 564)]
        bg1_patches = []
        bg2_patches = []

        for flo in leaf_locations:
            bg1_patches.append(image1_gray[flo[0]:flo[0] + PATCH_SIZE, 
                                            flo[1]:flo[1] + PATCH_SIZE])
            bg2_patches.append(image2_gray[flo[0]:flo[0] + PATCH_SIZE, 
                                            flo[1]:flo[1] + PATCH_SIZE])


        # Compute GLCM properties of each patch
        xs1 = []
        ys1 = []
        
        xs2 = []
        ys2 = []

        for patch in (image1_patches + bg1_patches):
            glcm = graycomatrix(image1_gray, distances=[5], angles=[0], levels=256, 
                                symmetric=True, normed=True)

            xs1.append(graycoprops(glcm, 'dissimilarity')[0, 0])
            ys1.append(graycoprops(glcm, 'correlation')[0, 0])
            
        for patch in (image2_patches + bg2_patches):
            glcm = graycomatrix(image2_gray, distances=[5], angles=[0], levels=256, 
                                symmetric=True, normed=True)
            
            xs2.append(graycoprops(glcm, 'dissimilarity')[0, 0])
            ys2.append(graycoprops(glcm, 'correlation')[0, 0])
            
        glcm1 = np.concatenate([xs1, ys1])
        glcm2 = np.concatenate([xs2, ys2])
        
        glcm_dist = self.euclideanDistance(glcm1, glcm2)
        
        return glcm_dist
        
        
    
    def hogDist(self):
        
        image1_gray = cv2.cvtColor(self.image1, cv2.COLOR_BGR2GRAY)
        image2_gray = cv2.cvtColor(self.image2, cv2.COLOR_BGR2GRAY)

        (H1, hogImage1) = feature.hog(image1_gray, orientations=9, pixels_per_cell=(8, 8),
                            cells_per_block=(2, 2), transform_sqrt=True, block_norm="L1",
                            visualize=True)
        (H2, hogImage2) = feature.hog(image2_gray, orientations=9, pixels_per_cell=(8, 8),
                            cells_per_block=(2, 2), transform_sqrt=True, block_norm="L1",
                            visualize=True)

        hogImage1 = exposure.rescale_intensity(hogImage1, out_range=(0, 255))
        hogImage1 = hogImage1.astype("uint8").flatten()
        
        hogImage2 = exposure.rescale_intensity(hogImage2, out_range=(0, 255))
        hogImage2 = hogImage2.astype("uint8").flatten()
        
        
        hd = self.euclideanDistance(hogImage1, hogImage2)
        
        return hd
    
    def lbpDist(self, a, b):
        
        desc = LocalBinaryPatterns(a, b)
        

        #Convert image to grey 
        image1_gray = cv2.cvtColor(self.image1, cv2.COLOR_BGR2GRAY)
        image2_gray = cv2.cvtColor(self.image2, cv2.COLOR_BGR2GRAY)
        
        hist1 = desc.describe(image1_gray)
        hist2 = desc.describe(image2_gray)
        
        ld = self.euclideanDistance(hist1, hist2)
        
        return ld
        
        
    
    def euclideanDistance(self, x, y):
        
        return sqrt(sum(pow(a-b,2) for a, b in zip(x, y)))
    
        
#Load Images       
flower1 = cv2.imread("img/image_00916.jpg")
flower2 = cv2.imread("img/image_01228.jpg")

#Apply the Distance Calculator class
lo = FeatureDistCalc(flower1, flower2)


## Calculate Late Fusion ##
# Initiatialize the Distance Calculator using the various features

#Color Moment
cmd = lo.colorMomentDist()

#Linear Binary Processing
lbpd = lo.lbpDist(24, 8)

#Color Histogram
chd = lo.colorHistogramDist()

#HOG
hd = lo.hogDist()

#GLCM
glcm = lo.glcmDist()

# Calculate late fusion
lf = (0.4*chd) + (0.3*lbpd) + (0.15*hd) + (0.1*cmd) + (0.5*glcm) 

#Display late fusion value
print(lf)
```

##### COSINE SIMILARITY

```python
from math import*
import cv2
```

```python
def square_rooted(x):
 
    return round(sqrt(sum([a*a for a in x])),3)
 
def cosine_similarity(x,y):
    
    numerator = sum(a*b for a,b in zip(x,y))
    denominator = square_rooted(x)*square_rooted(y)
    return round(numerator/float(denominator),3)

img5 = cv2.imread("img_query/image_01284.jpg")
img6 = cv2.imread("img_query/image_01289.jpg")

img5_flattened = img5.flatten()
img6_flattened = img6.flatten()

print(cosine_similarity(img5_flattened,img6_flattened))
```

### Content-Based Image Retrieval System (CBIR)

##### COLOR DESCRIPTOR

```python
import numpy as np
import cv2
import imutils
```

```python
class ColorDescriptor:
    def __init__(self, bins):
        # store the number of bins for the 3D histogram
        self.bins = bins    
        
    def describe(self, image):
        # convert the image to the HSV color space and initialize
        # the features used to quantify the image
        image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
        features = []
        
        # grab the dimensions and compute the center of the image
        (h, w) = image.shape[:2]
        (cX, cY) = (int(w * 0.5), int(h * 0.5))
        
        #create region based descriptor(top-left,
        # top-right, bottom-right, bottom-left)
        segments = [(0, cX, 0, cY), (cX, w, 0, cY), (cX, w, cY, h), (0, cX, cY, h)]
        
        
        # construct an elliptical mask representing the center of the
        # image
        (axesX, axesY) = (int(w * 0.75) // 2, int(h * 0.75) // 2)
        ellipMask = np.zeros(image.shape[:2], dtype = "uint8")
        cv2.ellipse(ellipMask, (cX, cY), (axesX, axesY), 0, 0, 360, 255, -1)
        
        # loop over the segments
        for (startX, endX, startY, endY) in segments:
            # construct a mask for each corner of the image, subtracting
            # the elliptical center from it
            cornerMask = np.zeros(image.shape[:2], dtype = "uint8")
            cv2.rectangle(cornerMask, (startX, startY), (endX, endY), 255, -1)
            cornerMask = cv2.subtract(cornerMask, ellipMask)
            
            # extract a color histogram from the image, then update the
            # feature vector
            hist = self.histogram(image, cornerMask)
            features.extend(hist)
            
        # extract a color histogram from the elliptical region and
        # update the feature vector
        hist = self.histogram(image, ellipMask)
        features.extend(hist)
        
        # return the feature vector
        return features

    def histogram(self, image, mask):
        # extract a 3D color histogram from the masked region of the
        # image, using the supplied number of bins per channel
        hist = cv2.calcHist([image], [0, 1, 2], mask, self.bins,
            [0, 180, 0, 256, 0, 256])
        
        # normalize the histogram if we are using OpenCV 2.4
        if imutils.is_cv2():
            hist = cv2.normalize(hist).flatten()
            
        # otherwise handle for OpenCV 3+
        else:
            hist = cv2.normalize(hist, hist).flatten()
            
        # return the histogram
        return hist
```

```python
class Searcher: 
    def __init__(self, indexPath):
        self.indexPath = indexPath

    def search(self, queryFeatures, limit=10):
        #initialize dictionary of results
        results ={}

        # open the index file for reading
        with open(self.indexPath) as f:
            # initialize the CSV reader
            reader = csv.reader(f)

            # loop over the rows in the index
            for row in reader:
                # parse out the image ID and features, then compute the
                # chi-squared distance between the features in our index
                # and our query features
                features = [float(x) for x in row[1:]]
                d = self.chi2_distance(features, queryFeatures)

                # now that we have the distance between the two feature
                # vectors, we can udpate the results dictionary -- the
                # key is the current image ID in the index and the
                # value is the distance we just computed, representing
                # how 'similar' the image in the index is to our query
                results[row[0]] = d

            # close the reader
            f.close()

        # sort our results, so that the smaller distances (i.e. the
        # more relevant images are at the front of the list)
        results = sorted([(v, k) for (k, v) in results.items()])
        
        # return our (limited) results
        return results[:limit]
    
    def chi2_distance(self, histA, histB, eps = 1e-10):
        # compute the chi-squared distance
        d = 0.5 * np.sum([((a - b) ** 2) / (a + b + eps)
            for (a, b) in zip(histA, histB)])
        # return the chi-squared distance
        return d
```
